# screen的使用


## 关于开启

```python
screen -S <窗口名称>
```

屏幕会闪一下，然后进入一个全新的黑色终端。这就是你的“后台”。在这里运行的代码，不受电脑关机、断网的影响。

现在的任务就是找到我们的python文件，可以使用locate或者是find，文件名比如是cv_project.py,然后准备运行

```python
locate cv_project.py
```

```python
find / -name "cv_project.py" 2>/dev/null
```

比如我们会看到下面的格式

```python
/tmp/pycharm_project_328/cv_project.py
```

然后此时就是我们先激活环境，然后进入到文件目录（可以在文件目录处用ls来查看是否在其中）

```python
conda activate base
```

```python
cd /tmp/pycharm_project_328/
```

然后此时如果已经有了py文件，那么直接运行就可以

```python
python cv_project.py
```

观察：看到终端打印出 Using device: cuda 和 Epoch [1/20] 之后，再进行下一步，此时也可以看看是否有问题报错

常见的报错是：

1.溢出（GPU不够用：RuntimeError: CUDA out of memory）

2.GPU比例太低，完全可以将batchsize调整高一些，可以采用再开一个终端然后输入下面指令



```python
nvidia-smi
```

然后我们进入最关键的一步，进行分离，这一步完成后就可以后端运行

在键盘上依次按下：

Ctrl + a （按住不松）

d （轻点一下，然后全部松开）

当你看到终端显示 [detached from 任务名称] 时，就已经完成挂机

## 重新连接与退出

现在需要解决的是我如何进入这个screen，然后关掉这个screen，首先我们先看看有哪些screen

```python
screen -ls
```

常规的进入与强制进入都如下

```python
screen -r "name"
screen -d -r "name"
```

如果此时我们需要让它停下，就使用control + c，进入KeyboardInterrupt 报错状态

谨慎使用：如果 control + c 没反应，直接关掉窗口（exit），或者在新终端输入 fuser -v /dev/nvidia0 找到进程 PID，然后 kill -9 <PID>，如果kill将会停止所有python进程

如果回到了光标闪烁等待输入的模式，现在我们就可以安全的退出

```python
exit
```

会看到 [screen is terminating]，随后自动回到主终端。此时这个窗口已经彻底消失。

最后的检查：输入 screen -ls，确保列表里没有多余的任务，同时打开nvidia-smi查看任务的状态
